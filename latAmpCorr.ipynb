### Import dependencies for linear regression

```python
from sklearn import linear_model
from scipy import stats
```

### Test linear regression on an example serie couple 

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(rndisofullX, rndisofullY)
print("slope: %f    intercept: %f" % (slope, intercept))
print("r-squared: %f" % r_value**2)
print(p_value)
```

### Plot one given serie couple example and its linear regression, slope, Fisher test p-value

```python
fig2 = plt.figure(figsize=(12,12))
plt.plot(cpisosecX, cpisosecY,markersize = 15,marker=".",ls= "", color = 'red', label = 'CP-ISO')
plt.plot(cpisosecX, intercept + slope*cpisosecX, 'b', label='fitted line, R = '+ str(r_value) +', Fisher test  ' +str(p_value))

plt.title('Amplitude gain function of latency advance - SECTOR', fontsize=20)
plt.xlabel('Latency advance (ms)', fontsize=20)
plt.ylabel('Amplitude gain (% test center peak)', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend()
plt.xlim(-15,30)
plt.ylim(-0.6,0.9)
```

```python
fig2 = plt.figure(figsize=(12,12))
plt.plot(rndisofullX, rndisofullY,markersize = 15,marker=".",ls= "", color = 'blue', label = 'RND-ISO')
plt.plot(rndisofullX, intercept + slope*rndisofullX, 'b', label='fitted line, R = '+ str(r_value) +', Fisher test  ' +str(p_value))

plt.title('Amplitude gain function of latency advance - FULL', fontsize=20)
plt.xlabel('Latency advance (ms)', fontsize=20)
plt.ylabel('Amplitude gain (% test center peak)', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend()
plt.xlim(-15,30)
plt.ylim(-0.6,0.9)
```

### Identify outliers

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(cpisofullX, cpisofullY)
print("slope: %f    intercept: %f" % (slope, intercept))
print("r-squared: %f" % r_value**2)

fig2 = plt.figure(figsize=(12,12))
plt.plot(cpisofullX, cpisofullY,markersize = 15,marker=".",ls= "", color = 'red', label = 'data points')
plt.plot(cpisofullX, intercept + slope*cpisofullX, 'b', label='fitted line, R squared  = 0.37')


#plt.scatter(cpisosecX, cpisosecY,s= 15, color = 'red')

plt.title('Amplitude gain function of latency advance - CP-ISO FULL', fontsize=20)
plt.xlabel('Latency advance (ms)', fontsize=20)
plt.ylabel('Amplitude gain (%)', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend()
```

```python
cpisoFull2_x   = np.loadtxt(cgwdir + 'cp_iso_Dlat50_full2.txt')
cpisoFull2_y   = np.loadtxt(cgwdir + 'cp_iso_Dres50_full2.txt')
```

### Linear regression correlation has indeed increased

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(cpisofull2X, cpisofull2Y)
print("slope: %f    intercept: %f" % (slope, intercept))
print("r-squared: %f" % r_value**2)

fig2 = plt.figure(figsize=(12,12))
plt.plot(cpisofull2X, cpisofull2Y,markersize = 15,marker=".",ls= "", color = 'red', label = 'data points')
plt.plot(cpisofull2X, intercept + slope*cpisofull2X, 'b', label='fitted line, R squared = 0.67')
plt.title('Amplitude gain function of latency advance - CP-ISO FULL', fontsize=20)
plt.xlabel('Latency advance (ms)', fontsize=20)
plt.ylabel('Amplitude gain (%)', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend()
```
